# ECRH scoping

This notebook is for scoping the ECRH frequency and power needed in the mirror

```python
import numpy as np
import scipy.constants as const
```

```python
# Mirror parameters
m_i = 2.5*const.atomic_mass
m_e = const.m_e

E_b_keV = 45 # [keV]
Te_keV = 0.1*E_b_keV
```

## Critical energy (50-50 split of beams slowing down on ions vs electrons)

If $E_b < E_c$, then the beams primary slow down on the ions.

```python
# See Freidberg, Fusion Energy, Eq. 9.67, assuming m_b = m_i
Ec_keV = ( 3*np.sqrt(np.pi)/2 ) ** (2/3)
Ec_keV *= Te_keV * (m_i/m_e)**(1/3)
print(f"Critical Energy = {Ec_keV:.1f} keV")
```

## Estimate for required ECRH power

The electron temperature must be high enough so that the confinement time scales with the ion-ion collsion rate, not the slowing down rate of ions on electrons. From Egedal22, this means that $T_e \geq 0.1 E_b$.
